```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns
import plotly.offline as iplot
import cufflinks as cf
cf.go_offline()
```

In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
data_frame=sns.load_dataset('iris')
```

```python
data_frame.head()
```

```python
data_frame=data_frame[['sepal_length','sepal_width','petal_width']]
```

```python
data_frame.iplot(kind='bubble3d',x='sepal_length',y='sepal_width',z='petal_width',xTitle='sepal_length',yTitle='sepal_width',zTitle='petal_width',size='petal_width',colors=['green','blue','red'])
```
